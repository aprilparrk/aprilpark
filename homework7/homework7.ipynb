```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def make_sine_wave(x, A, w):
    return A * np.sin(w * x)
    
```

```python
x = np.linspace(0, 2 * np.pi, 1000)
amplitude = np.array([0.5, 1, 1.5, 2, 2.5])
frequency = np.array([1, 2, 3, 4, 5])

```

```python
fig, axes = plt.subplots(1, 3, figsize=(25, 5))
for A in zip(amplitude):
    y = make_sine_wave(x, A, 1)
    axes[0].plot(x, y, label=f"A={A}, w=1")

for w in zip(frequency):
    y = make_sine_wave(x, 1, w)
    axes[1].plot(x, y,label=f"A=1, w={w}")

for A, w in zip(amplitude, frequency):
    y = make_sine_wave(x, A, w)
    axes[2].plot(x, y, label=f"A={A}, w={w}")
titles = [
    "Varying Amplitude",
    "Varying Frequency",
    "Varying Amplitude and Frequency Together"
]

for ax, title in zip(axes, titles):
    ax.set_title(title, fontsize=14)
    ax.set_xlabel("x (radians)", fontsize=14)
    ax.set_ylabel("y = Asin(x)", fontsize=14)
    ax.legend(fontsize=12)   

plt.show()
```
