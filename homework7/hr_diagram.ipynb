```python
import numpy as np
import matplotlib.pyplot as plt
data = np.loadtxt('../../course_assignments/homework7/star_data.txt', skiprows=1)
data.shape #2642 rows
```

```python
magnitude = data[:,0]
temperature = data[:,1]
```

```python
print(np.min(magnitude)) #brightest star
print(np.max(magnitude)) #dimmest star
print(np.min(temperature)) #hottest star
print(np.max(temperature)) #coldest star
```

```python
brightest = np.argmin(magnitude)
dimmest = np.argmax(magnitude)
hottest = np.argmax(temperature)
coldest = np.argmin(temperature)
```

```python
log_magnitude = np.log(magnitude)
log_temperature = np.log(temperature)
plt.figure(figsize=(8,6))

plt.scatter(log_temperature, log_magnitude, label='Stars', s=30)
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()

plt.scatter(log_temperature[brightest], log_magnitude[brightest], 
            color='black',marker='*',s=250, label='Brightest Star')
plt.scatter(log_temperature[dimmest], log_magnitude[dimmest], 
            color='purple',marker='*', s=250,label='Dimmest Star')
plt.scatter(log_temperature[hottest], log_magnitude[hottest], 
            color='green',marker='*',s=250,label='Hottest Star')
plt.scatter(log_temperature[coldest], log_magnitude[coldest], 
            color='orange',marker='*',s=250,label='Coldest Star')



plt.xlabel("Temperature (log K)", fontsize=14)
plt.ylabel("Magnitude", fontsize=14)
plt.title("Hertzsprung-Russell Diagram", fontsize=18)
plt.legend(fontsize=9)

plt.show()
```
