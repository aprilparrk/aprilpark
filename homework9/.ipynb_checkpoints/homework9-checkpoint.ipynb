```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
df = pd.read_csv("../../course_assignments/homework9/noisy_curvefit_data.csv")
x = df["x"].values
y_noise_1 = df["y_noise_1"].values
```

```python
plt.figure(figsize=(8,8))
plt.scatter(df["x"], df["y_noise_1"], color="gray", label = "noisy data")

plt.xlabel("x")
plt.ylabel("y noise 1")
plt.title("noisy curve fit")
plt.legend()
plt.show()



```

```python
def model(x,A,k):
    return A * np.exp(-k * x)
y_noise_1 = df["y_noise_1"].values
p0 = [1, 1]
popt, pcov = curve_fit(model, x, y_noise_1, p0=p0)
A_fit, k_fit = popt
print(f"Fit: y = {A_fit:.2f} * e^(-{k_fit:.2f} * x)")

residuals = y_noise_1 - model(x, A_fit, k_fit)
chi2 = np.sum(residuals**2)
dof = len(x) - len(popt)
chi2_reduced = chi2 / dof

plt.figure(figsize=(8,8))
plt.scatter(x, y_noise_1, color="gray", label="noisy data")
plt.plot(x, model(x, A_fit, k_fit), color="red", label="fitted model")
plt.plot(x, model(x, 5, 0.8), linestyle="--", color="black", label="true model")

plt.xlabel("x")
plt.ylabel("y")
plt.title(f"noise level 1,{chi2_reduced:.2f}")
plt.legend()
plt.show()
```

```python
def fit_model(x, y):
    p0 = [1, 1]
    popt, pcov = curve_fit(model, x, y, p0=p0)
    A_fit, k_fit = popt
    
    residuals = y - model(x, A_fit, k_fit)
    chi2 = np.sum(residuals**2)
    dof = len(x) - len(popt)
    chi2_red = chi2 / dof
    
    return A_fit, k_fit, chi2_red
fig, axes = plt.subplots(2, 2, figsize=(12,10))
axes = axes.flatten()
noise_cols = ["y_noise_2", "y_noise_3", "y_noise_4", "y_noise_5"]

results = [] 
results.append([
    "y_noise_1",
    A_fit,          
    k_fit,
    chi2_reduced
])
for i, col in enumerate(noise_cols):
    y = df[col].values
    
    A_fit, k_fit, chi2_red = fit_model(x, y)
    results.append([col, A_fit, k_fit, chi2_red])
    
    ax = axes[i]
    ax.scatter(x, y, color="gray", label="noisy data")
    ax.plot(x, model(x, A_fit, k_fit), color="red", label="fitted model")  
    ax.plot(x, model(x, 5, 0.8), linestyle="--", color="black", label="True Model")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_title(f"{col},{chi2_red:.2f}")
    ax.legend()



    
plt.tight_layout()
plt.show()

```

```python
#As the noise level increases, the red and black dotted lines become further from each other. Higher noise makes the line of best fit less accurate.
#reduced chi-squared value increases as noise level increases, and produces worse fit
```

```python
df_new = pd.DataFrame(results, columns=["Noise Level", "A_fit", "k_fit", "chi2_reduced"])
df_new
```

```python
df_new.to_csv("noisy_parameters.csv", index=False)
```
