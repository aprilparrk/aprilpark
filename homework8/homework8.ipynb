```python
latmed = df["latitude"].median()
latmed
```

```python
longmed = df["longitude"].median()
longmed
```

```python
df["latdis"] = (df["latitude"] - latmed).abs()
df["longdis"] = (df["longitude"] - longmed).abs()
```

```python
df.shape
```

```python
#52 rows, 6 columns
```

```python
closestlat = df.loc[df["latdis"].idxmin()]
print(f"{closestlat['state']}, {closestlat['latdis']:.2f}")
```

```python
farthestlat = df.loc[df["latdis"].idxmax()]
print(f"{farthestlat['state']}, {farthestlat['latdis']:.2f}")
```

```python
closestlong = df.loc[df["longdis"].idxmin()]
print(f"{closestlong['state']}, {closestlong['longdis']:.2f}")
```

```python
farthestlong = df.loc[df["longdis"].idxmax()]
print(f"{farthestlong['state']}, {farthestlong['longdis']:.2f}")
```

```python
df2 = sns.load_dataset("penguins")
df2
```

```python
df2_dropped = df2.dropna(axis= 1)
df2_dropped
```

```python
df2_dropped.shape
#344 rows, 2 columns
```

```python
df2_dropped.columns
#species: the type/sort of penguines
#island: habitat of the penguin
#bill_length: length of bills
#bill_depth: measurement from the top of beak to lower edge at the thickest point
#flipper_length: length of modified wing that is adapted for swiwmming
#body_mass: total mass of the penguin
#sex: male/female
```

```python
fig, axs = plt.subplots(2, 2)
axs[0][0].scatter(df2["bill_length_mm"], df2["bill_depth_mm"])
axs[0][0].set_title("bill length vs depth")
axs[0][0].set_xlabel("bill length(mm)")
axs[0][0].set_ylabel("bill depth(mm)")


axs[0][1].scatter(df2["flipper_length_mm"], df2["body_mass_g"])
axs[0][1].set_title("flipper length vs body mass")
axs[0][1].set_xlabel("flipper length(mm)")
axs[0][1].set_ylabel("body mass(g)")

sns.histplot(df2["bill_length_mm"], ax=axs[1][0])
axs[1][0].set_title("bill length")

sns.boxplot(x=df2["species"], y=df2["body_mass_g"], ax=axs[1][1])
axs[1][1].set_title("body mass by species")
axs[1][1].set_xlabel("species")
axs[1][1].set_ylabel("body mass(g)")

axs[0][0].scatter(df2["bill_length_mm"], df2["bill_depth_mm"], color="red", label="Bill")
axs[0][0].legend()

sns.histplot(df2["bill_length_mm"], ax=axs[1][0], color="green")
sns.boxplot(x=df2["species"], y=df2["body_mass_g"], ax=axs[1][1], color="purple")



plt.tight_layout()
plt.show()
```
