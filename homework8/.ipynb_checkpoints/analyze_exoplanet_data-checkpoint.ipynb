```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../../aprilpark/homework8/exoplanet_data.csv", comment = '#')
```

```python
df.shape #There are 570 rows and 92 columns
```

```python
print(df.columns) 
#"sy_snum", "sy_pnum", "rastr", "sy_kmag", "st_teff"
```

```python
#Planet Name, pl_name: Planet name most commonly used in the literature, official name of the planet	
#Host Name, hostname: Stellar name most commonly used in the literature, unique identifier used when computing the planets
#Discovery Method,discoverymethod: Method by which the planet was first identified, techniques used to find planets, normally indirect
#Spectral Type,st_spectype: Classification of the star based on their spectral characteristics following the Morgan-Keenan system, , classification system for stars determined by analyzing absorption lines in their light spectra	
#Planet Mass or Mass*sin(i) [Jupiter Mass], pl_bmassj: Amount of matter contained in the planet, measured in units of masses of Jupiter, total mass of Jupiter
#Equilibrium Temperature[K],pl_eqt: The equilibrium temperature of the planet as modeled by a black body heated only by its host star, or for directly imaged planets, the effective temperature of the planet required to match the measured luminosity if the planet were a black body, balance between energy absorbed and energy emitted
```

```python
small = [
    "pl_name","hostname","discoverymethod","st_spectype","pl_eqt","pl_bmassj"
]

smalldf = df[small]
smalldf = smalldf.dropna()
```

```python
smalldf.shape #There are 67 rows and 6 columns
```

```python
smalldf.head()
```

```python
smalldf.tail()
```

```python
print(smalldf.columns)
```

```python
print(smalldf["discoverymethod"].value_counts())
#methods: transit(52), radial velocity(13), imaging(2)
#The transit method was used the most, and imaging was used the least
```

```python
#Transit: technique for discovering exoplants by observing periodic dimming of a star's light as planet passes in front of it
#Radial Velocity: detects exoplanets by measuring the tiny woble of star caused by gravitational pull of an orbiting planet
#Imaging: Direct imaging, involving special equipments like coronagraphs and adaptive optics to block a star's light and capture faint light from the planet
```

```python
plt.figure(figsize=(8,6))

for method, subset in smalldf.groupby("discoverymethod"):
    plt.scatter(subset["pl_bmassj"], subset["pl_eqt"], label= method)
plt.xlabel("planet mass")
plt.ylabel("equilibrium temperature")
plt.title("planet mass vs equilibrium temperature")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
for method, subset in smalldf.groupby("discoverymethod"):
    plt.scatter(np.log(subset["pl_bmassj"]), np.log(subset["pl_eqt"]),
                label=method)
plt.xlabel("log planet mass")
plt.ylabel("log equilibrium temperature")
plt.title("log planet mass vs equilibrium temperature")
plt.legend()
plt.show()
#Exoplanets found by imaging are more likely to be massive




massive = smalldf[smalldf["pl_bmassj"] > 5]
coldest = massive.loc[massive["pl_eqt"].idxmin()]


print(coldest.name)
print(coldest['pl_name'])
print(coldest['pl_eqt'])
print(coldest['pl_bmassj'])
print(coldest['st_spectype'])
```

```python
plt.figure(figsize=(8, 6))
for method, subset in smalldf.groupby('discoverymethod'):
    plt.scatter(np.log(subset["pl_bmassj"]), np.log(subset["pl_eqt"]), label = method)

plt.scatter(coldest["pl_bmassj"], coldest["pl_eqt"],
            color='red', marker='x', s=100, label=f"Coldest planet: {coldest['pl_name']}")

plt.xlabel("log planet mass")
plt.ylabel("log equilibrium temperature")
plt.title("log planet mass vs Equilibrium temperature")
plt.legend()
plt.show()
```
